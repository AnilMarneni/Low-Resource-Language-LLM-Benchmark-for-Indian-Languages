# LLM Indic Language Benchmark Analysis

This notebook allows researchers to quickly load the aggregated `.csv` benchmark results from the framework to perform custom clustering, pivot tables, and statistical significance testing.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Find latest benchmark run
results_dir = "../results/benchmarks"
csv_files = [f for f in os.listdir(results_dir) if f.endswith(".csv")]
csv_files.sort(reverse=True)

if not csv_files:
    print("No benchmark results found. Run the benchmark tool first.")
else:
    latest_file = os.path.join(results_dir, csv_files[0])
    print(f"Loading {latest_file}")
    df = pd.read_csv(latest_file)
    display(df.head())
```

### Model Performance by Task (Averaged Over Languages)

```python
if 'df' in locals():
    # Aggregate metrics to compute mean scores if applicable
    numeric_cols = df.select_dtypes(include='number').columns
    pivot = df.pivot_table(index='Model', columns='Task', values=numeric_cols, aggfunc='mean')
    display(pivot)
```
